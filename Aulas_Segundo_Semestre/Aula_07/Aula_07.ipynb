Crie um programa que solicita o salário de um funcionário. 
Após isso, crie uma segunda função para calcular o desconto do funcionário.
Posteriormente, crie uma terceira função para retornor o sálario que o funcionara deverá receber.

ate 2000 com 10% de desconto
2000 até 4800 com 15% de desconto
4800 ate 6000 com 20% de desconto
Acima de 6000 com 30% de desconto

```python

salario = float(input('Digite o valor do seu salário: '))
    

def desconto(salario):
    if salario <= 2000:
        desconto =  salario * 0.1

    elif 2000 < salario <= 4800:
        desconto = salario * 0.15

    elif 4800 < salario <= 6000:
        desconto = salario * 0.2

    else:
        desconto = salario * 0.3
    return desconto

def salario_liquido(salario):
    salario_liquido = salario - desconto(salario) 
    return salario_liquido

print('Desconto: ', desconto(salario))
print('Salario liquido: ', salario_liquido(salario))
    
```

crie um programa para calcular a area de um circulo, uma função para calcular a área e outra pra classificar a área

area > 100 - grande
area > 50 - medio
else - pequeno

```python
# Definindo a area do circulo
def circulo():
    r = float(input('Defina o raio de um circulo: '))
    pi = 3.1415
    area = pi * r**2
    return area

# Classificando o circulo
def classificar():
    area = circulo()
    if area > 100:
        return print('Circulo grande', area)
    elif area > 50:
        return print('Circulo médio', area)
    else:
        return print('Circulo pequeno', area)

classificar()
```

Crie um função para multiplicar os valores de duas listas.
A primeira função faz a multplicação e a segunda pegará a primeira e ultilizará um append prdutos = [] para multiplicar os respectivos índices das duas listas

lista 1 = [1, 2, 3, 4]
lista 2 = [5, 6, 7, 8]

```python
lista_1 = [1, 2, 3, 4]
lista_2 = [5, 6, 7, 8]

def multplicar(y, z):
    x = y * z
    return x


def multplicar_lista(lista1, lista2):
    produto = []
    for i in range(len(lista1)):
        produto.append(multplicar(lista1[i], lista2[i]))
    return produto
multplicar_lista(lista_1, lista_2)
```

Crie um programa para uma calculadora simples de tal forma que tenha uma função para cadâ operação (-, +, /, *).
Posteriormente, crie uma função para operações que fará as quatros ao mesmo tempo

```python
a = float(input('Defina o valor de A: '))
b = float(input('Defina o valor de B: '))

# Criando as operações
def soma(a, b):
    return a + b
def sub(a, b):
    return a - b
def div(a, b):
    return a / b
def mult(a, b):
    return a * b

def calcular(operacoes):
    resultado = []
    for op, a, b in operacoes:
        if op == '+':
            resultado.append(soma(a, b))
        elif op == '-':
            resultado.append(sub(a, b))
        elif op == '/':
            resultado.append(div(a, b))
        else:
            resultado.append(mult(a, b))
    return resultado

expressao = [('+', a, b), ('-', a, b), ('/', a, b), ('*', a, b)]    
resultado_final = calcular(expressao)
print(resultado_final)
```
