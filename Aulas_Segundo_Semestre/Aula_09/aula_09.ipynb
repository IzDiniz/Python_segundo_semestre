```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame([[4, 9, 2]] * 4, columns=['A', 'B', 'C'])
df
```

```python
# Raiz Quadrada

# Apply permite que você some as colunas ou linhas do DataFrame
# O parâmetro Axis serve para selecionar se a gente vai contar as colunas ou linhas, com os parâmetros 0 ou 1
df.apply(np.sqrt)
soma_linha = df.apply(np.sum, axis = 1)
soma_coluna = df.apply(np.sum, axis = 0)
soma_coluna
```

```python
# juntar função apply com lambda

# Queremos somar toda a coluna 'C' mas criando uma nova coluna 'D', ultilizando lambda, usando o lambda dentro do aplly
# para multiplicar todos os valores do C por 2, 'x' do lambda é cada linha do 'C' multplicado por 2, independente da quantidade 
df['D'] = df['C'].apply(lambda x: x*2)
df
```

```python
# Somar colunas e linhas 

# Row - É um metôdo do 'pandas' usado para pegar as linhas de uma tabela
resultado = df.apply(lambda row: row['A'] + row['B'], axis= 1)
resultado
```

```python
# Calcular a média, mediana, min, max

# DataFrame a partir de um dicionário
df = pd.DataFrame({
    'A': [6, 5, 8, 11],
    'B': [23, 20, 5, 30]})
df
```

```python
resultado_media_a = df['A'].apply(lambda x: x).mean()
print(resultado_media_a)

resultado_mediana_a = df['A'].apply(lambda x: x).median()
print(resultado_mediana_a)
```

```python
df_nomes = pd.DataFrame({
    'Nome': ['João', 'Ana', 'João', 'Pedro']})

df_nomes['Nome'] = df_nomes['Nome'].apply(lambda x: x + ' Silva')
df_nomes 
```

```python
# DataSet sobre falhas e maquinas
falhas = pd.read_csv('manutencao_preditiva.csv')
falhas
```

```python
# 1 - Crie um programa para somar as variáveis temperatura do ar e processo e retorne uma nova variável
falhas['Coluna nova'] = falhas.apply(lambda row: row['Temperatura Ar [K]'] + row['Temperatura Processo [K]'], axis= 1)
falhas
```

```python
import matplotlib.pyplot as plt

# Normalização de dados

df_sample = pd.DataFrame({
    'A':[6, 5, 8, 11],
    'B': [100, 122, 156, 189],
    'indice': [1, 2, 3, 4]})

x = df_sample['indice']
y1 = df_sample['A']
y2 = df_sample['B']

plt.plot(x, y1, x, y2)
plt.show()
```

```python
df_normalizando = df_sample.apply(lambda x: (x - x.min()) / (x.max() - x.min()))
df_normalizando
```

```python
df = pd.DataFrame({
    'Texto': ['Olá', 'Mundo', 'Python', 'Pandas']
})

df['contagem'] = df['Texto'].apply(lambda texto: len(texto))
df
```

```python
# Criar uma condição para inserir uma coluna chamada comprimento_texto e se o comprimento
# for maior do que 5, insira longo, caso contrário curto.

df['comprimento_texto'] = df['Texto'].apply(lambda texto: 'Longo' if len(texto) > 5 else 'Curto')
df
```

```python
# Crie um programa para verificar o comprimento do texto na coluna tipo da falha. Após isso, insira
# uma nova coluna mostrando. se o comprimento menor ou igual a 10 - curto. se o comprimento maior que 10
# e menor que 15 - médio. se o comprimento maior que 15 - longo.

dados = pd.read_csv('manutencao_preditiva.csv')
dados['verificando comprimento'] = dados['Tipo da Falha'].apply(lambda texto: 'curto' if len(texto) <=10 else 'Médio' if len(texto) > 10 and len(texto) >= 15 else 'Longo')
dados['verificando comprimento'].value_counts()
```

```python
df_new = pd.DataFrame({
    'Var_1': ['A', 'A', 'B', 'A', 'C', 'C', 'A', 'B', 'A', 'B'],
    'Var_2': [100, 122, 156, 189, 50, 63, 9, 15, 78, 10]
})

df_new
```

```python
# pega cada variável com a mesma categoria, somando e depois dividindo pela quantidades de variáveis
agrupado_media = df_new.groupby('Var_1').mean().reset_index()
agrupado_mediana = df_new.groupby('Var_1').median().reset_index()
agrupado_minimo = df_new.groupby('Var_1').min().reset_index()
agrupado_maximo = df_new.groupby('Var_1').max().reset_index()

print(f'O valor da média é: {agrupado_media}')
print(f'O valor da médiana é: {agrupado_mediana}')
print(f'O valor do mínimo é: {agrupado_minimo}')
print(f'O valor da máxima é: {agrupado_maximo}')
```

```python
agrupando_media_sem_reset = agrupado_media = df_new.groupby('Var_1').mean()
print(agrupando_media_sem_reset)
```
