Calculando a Média da uma lista


```python
#calcule a média de uma lista de números

def calcular_media(lista):
    return sum(lista) / len(lista)
# Somando os numeros da lista e depois dividindo pelos índices da lista

numeros = [10, 20, 30, 40, 50]
media = calcular_media(numeros)
print('A média dos números é: ', media)
```

Encontrar o maior valor de uma Lista

```python
# Encontrar maior valor de uma lista de números
def maior_funcao(lista):
    return max(lista)

numeros = [10, 20, 30, 40, 50]
media = maior_funcao(numeros)
print('O maior número da lista é: ', media)
```

Econtrando e separando os numeros positivos e negativos

```python
#Minha resolução - Verificando se os elementos de uma lista são positivos
def positivos(lista):
    return [x for x in lista if x > 0], [y for y in lista if y < 0]

numeros = [-10, 20, 30, 40, -50]
num = positivos(numeros)
print('Os números positivos e negativos dessa lista são: ', num)
         
```

```python
#Resolução do professor - Verificar se os elementos são positivos
def positivos_negativos(lista):
    positivos = [num for num in lista if num > 0]
    negativos = [num for num in lista if num < 0]
    return positivos, negativos

numeros = [10, -20, 30, -40, 50, -60, 70]
positivos, negativos = positivos_negativos(numeros)
print('Numeros negativos: ', negativos)
print('Numeros positivos: ', positivos)
```

Retornando a Soma dos elementos e a Média de uma lista 

```python
# Retornando média e a soma da lista 
def media_soma(lista):
    soma = sum(lista)
    media = sum(lista) / len(lista)
    return soma, media

numeros = [10, 0, 7, 10, 22, 5]

soma, media = media_soma(numeros)
print('Soma dos elementos: ',soma)
print('Média dos elementos: ',media)
```

Retornando o min e o max da lista usando função def


```python
# Retornando o minimo e o maximo da lista 
def maximo_minimo(lista):
    maximo = max(lista)
    minimo = min(lista)
    return maximo, minimo

numeros = [10, 0, 7, 10, 22, 5]

maximo, minimo = maximo_minimo(numeros)
print('Maximo dos elementos: ', maximo)
print('Minimo dos elementos: ',minimo)
```

Calculando os valores da lista ao quadrado 

```python
#Calculando a soma dos numeros ao quadrado
def soma_quadrado(lista):
    return sum(num**2 for num in lista)

numeros = [10, 2, 7, 9, 10, 22, 5]
elevado = soma_quadrado(numeros)
print(elevado)
```

Verifícando quantas vezes um elemento aparece na lista

```python
def elementos(lista, elemento):
    return lista.count(elemento)

numeros = [10, 2, 7, 9, 10, 22, 5, 7]
elemento = 7
ocorrencia = elementos(numeros, elemento)
print('O elemento', elemento, 'aparece', ocorrencia, 'vezes')
```

Invertendo elementos de uma lista

```python
def inverter_lista(lista):
    return lista[::-1]

numeros = [1, 2, 3, 4, 5]
lista_invertida = inverter_lista(numeros)
print('Lista invertida', lista_invertida)
```

```python
def esta_lista_crescente(lista):
    return lista == sorted(lista)

numeros = [1, 2, 3, 4, 5]
esta_ordenada = esta_lista_crescente(numeros)
print(esta_ordenada)
```

removendo elementos duplicados

```python
def remover_duplicados(lista):
    return list(set(lista))

numeros = [1, 2, 3, 4, 5, 1, 2, 3]
sem_duplicados = remover_duplicados(numeros)
print(sem_duplicados)
```
