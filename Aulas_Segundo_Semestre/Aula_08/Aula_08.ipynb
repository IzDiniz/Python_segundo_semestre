```python

comeca_com_g = lambda x:True if x.startswith('G') else False
print(comeca_com_g('Gabriel'))
print(comeca_com_g('Adalto'))
```

Funo que cria funo

```python
def const_quadratica(a, b, c):
    return lambda x: a*x**2 + b*x + c
    
f = const_quadratica(2, 3, 2)
# Numero 5  o valor de 'X'
print(f(5))
```

Exerccios

```python
# 01 - Crie uma funo dentro de uma funo para calcular a mdia quadrada de 2 numeros.
def media_ponderada(peso1, peso2):
    return lambda x,y : ((x * peso1) + (y * peso2)) / (peso1 + peso2)
    
f = media_ponderada(2, 3)
print(f(2, 3))
```

```python
# 02 - Crie uma funo para calcular o perimetro de um retngulo com base na altura e largura.
def perimetro_retangulo(largura, altura):
    return lambda:2 * (largura + altura)

calcular_perimetro = perimetro_retangulo(5, 3)
resultado = calcular_perimetro()
resultado
```

```python
# 03 - Crie uma funo para calcular o volume de um retngulo.
def area_retangulo():
    return lambda x,a,b: a * b * x

area = area_retangulo()
print(area(2,2,2))
```

```python
# 04 - Crie uma funo para calcular a equao de Newton: F = M * a, dando dois parmetros M e A
def equacao_newton():
    return lambda m, a: m * a

f = equacao_newton()
print(f(2, 3))
```

Criando a area tringulo, quadrado, circulo e retangulo...


```python
# ** chama um argumento para separa o que vai ser dentro da funo, nesse caso, area do circulo, tringulo e etc..
def calculo_area(forma, **geometrica):
    area_funcoes = {
        "quadrado": lambda lado: lado **2,
        "retangulo": lambda base, altura: base * altura,
        "circulo": lambda raio: 3.1415 * raio **2,
        "triangulo": lambda base, altura: (base * altura) /2}
    if forma in area_funcoes:
        return area_funcoes[forma](**geometrica)
    else:
        return "Forma geometrica no identificada"
    
```

```python
print("Area do quadrado", calculo_area("quadrado", lado = 4))
print("Area do retangulo", calculo_area("retangulo", base = 4, altura = 5))
print("Area do circulo", calculo_area("circulo", raio = 3))
print("Area do triangulo", calculo_area("triangulo", base = 2, altura = 1))
```

```python
numeros = [2, 5, 7, 4]

# Ordena do menor para o mair
ordenar = sorted(numeros)
ordenar
```

```python
palavras = ['a', 'c', 'd', 'b']
ordenar2 = sorted(palavras)
ordenar2
```

```python
frutas = ['abacaxi', 'pera', 'laranja', 'uva']

# pelo comprimento das palavras
ordenar3 = sorted(frutas, key=lambda fruta: len(fruta))
print(ordenar3)
```

```python
# Ordenar uma listas de tuplas pelo primeiro ou segundo elemento, apenas verifica o elemento 0 primeiro, e caso haja repetidos ele verifica o outro indice

tuplas = [(2, 3), (2, 8), (4, 3), (2, 1)]
ordenar_tupla = sorted(tuplas, key=lambda tupla: tupla[0])
ordenar_tupla
```
